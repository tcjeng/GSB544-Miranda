It should also be clear by how `plotnine` treats these variables.

1. Create side-by-side boxplots of the `mpg` variable by the different values of the `am` variable. What happens if you convert the `am` variable back to float values and then try to create this same plot? Explain the differences between the two plots.

2. Create overlaid histograms of the `hp` variable for the different values of the `cyl` variable. What happens if you convert the `cyl` variable back to float values and then try to create this same plot? Explain the differences between the two plots.

```python
# Create plot for (1) here
from plotnine import ggplot, aes, geom_boxplot, labs

# Create boxplot where 'am' is categorical
plot1 = (
    ggplot(df, aes(x='am', y='mpg')) +
    geom_boxplot() +
    labs(
        title="Boxplot of MPG by AM (Categorical)",
        x="AM (Transmission)",
        y="MPG"
    )
)
print(plot1)

# Create boxplot where 'am' is numerical (float)
df['am'] = df['am'].astype(float)
plot2 = (
    ggplot(df, aes(x='am', y='mpg')) +
    geom_boxplot() +
    labs(
        title="Boxplot of MPG by AM (Numerical)",
        x="AM (Transmission)",
        y="MPG"
    )
)
print(plot2)
```

The first boxplot treats 'am' as a categorical variable, so there will be two distinct boxplots, one for each category of the transmission type (0 = automatic, 1 = manual). Since boxplots are built for categorical variables, when converted 'am' to numeric variables, it confuses the program.

```python
# Create plot for (2) here
from plotnine import ggplot, aes, geom_histogram, labs

# Ensure 'cyl' is treated as categorical
df['cyl'] = df['cyl'].astype(str)

# Create overlaid histograms where 'cyl' is categorical
plot1 = (
    ggplot(df, aes(x='hp', fill='cyl')) +  # Use 'fill' to differentiate by 'cyl'
    geom_histogram(position='identity', alpha=0.5, bins=30) +  # Overlaid histograms
    labs(
        title="Overlaid Histograms of HP by Cylinders (Categorical)",
        x="HP",
        y="Count"
    )
)
print(plot1)

# Create overlaid histograms where 'cyl' is treated as numeric
df['cyl'] = df['cyl'].astype(float)
plot2 = (
    ggplot(df, aes(x='hp', fill='factor(cyl)')) +  # 'factor(cyl)' treats 'cyl' as categorical
    geom_histogram(position='identity', alpha=0.5, bins=30) +  # Overlaid histograms
    labs(
        title="Overlaid Histograms of HP by Cylinders (Numerical)",
        x="HP",
        y="Count"
    )
)
print(plot2)
```

we need to add in factor(cyl) when 'cyl' is numeric value to force it to act like categorical values, otherwise, the histogram wouldn't make sense

Some of these variables, like the `cyl` variable, have numeric values that actually make sense as numbers (i.e. the number of cylinders in the engine). However, it doesn't make the most sense to "do math" with this type of variable (e.g. take averages and such) because there are so few different value this can take on AND they're an explicit choice made by the car manufacturer. So, it makes more sense to treat the `cyl` variable as a categorical variable despite it having numeric values.

You will need to keep these kinds of nuances about data in mind as you work with an increasing variety and richness of data, and do more complex things with them.

Choose 3 new `geometries` from the [data-to-viz website](https://www.data-to-viz.com/) for graphs that you'd like to explore using the `mtcars` dataset, and then create those graphs!

```python
# Create plot with first new geometry here
# density plot
from plotnine import ggplot, aes, geom_density, labs

# Ensure 'cyl' is treated as categorical
df['cyl'] = df['cyl'].astype(str)

# Create the density plot
plot = (
    ggplot(df, aes(x='hp', color='cyl', fill='cyl')) +  # 'cyl' as categorical mapped to color and fill
    geom_density(alpha=0.5) +  # Semi-transparent fill
    labs(
        title="Density Plot of HP by Cylinders (Categorical)",
        x="HP",
        y="Density"
    )
)
print(plot)
```

```python
# Create plot with second new geometry here
# scatter plot
from plotnine import ggplot, aes, geom_point, labs

# Ensure 'cyl' is treated as categorical
df['cyl'] = df['cyl'].astype(str)

# Create scatter plot with 'cyl' as categorical
plot1 = (
    ggplot(df, aes(x='hp', y='cyl', color='cyl')) +  # 'cyl' as categorical mapped to color
    geom_point() +
    labs(
        title="Scatter Plot of HP vs Cylinders (Categorical)",
        x="HP",
        y="Cylinders"
    )
)
print(plot1)
```

```python
# Create plot with third new geometry here
# area plot
from plotnine import ggplot, aes, geom_area, labs

# Ensure 'cyl' is treated as categorical
df['cyl'] = df['cyl'].astype(str)

# Create the area plot with 'cyl' as categorical
plot1 = (
    ggplot(df, aes(x='hp', fill='cyl')) +  # 'cyl' as categorical, mapped to fill
    geom_area(stat='density', alpha=0.6) +  # Fill the area under the density curve
    labs(
        title="Area Plot of HP by Cylinders (Categorical)",
        x="HP",
        y="Density"
    )
)
print(plot1)
```
